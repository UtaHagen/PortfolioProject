```python
import yfinance as yf
import pandas as pd
import numpy as np

# Define a function to calculate beta
def calculate_beta(stock_ticker, market_ticker, start_date, end_date):
    # Fetch historical data
    stock_data = yf.download(stock_ticker, start=start_date, end=end_date)
    market_data = yf.download(market_ticker, start=start_date, end=end_date)
    
    # Calculate daily returns
    stock_returns = stock_data['Adj Close'].pct_change().dropna()
    market_returns = market_data['Adj Close'].pct_change().dropna()
    
    # Covariance between stock and market returns
    covariance = np.cov(stock_returns, market_returns)[0, 1]
    
    # Variance of market returns
    market_variance = np.var(market_returns)
    
    # Calculate beta
    beta = covariance / market_variance
    return beta

# Example usage
sp500_ticker = '^GSPC'
start_date = '2020-01-01'
end_date = '2023-01-01'

# Fetch the list of S&P 500 tickers
sp500_tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()

# Calculate beta for each stock
betas = {}
for ticker in sp500_tickers:
    try:
        beta = calculate_beta(ticker, sp500_ticker, start_date, end_date)
        betas[ticker] = beta
    except Exception as e:
        betas[ticker] = None
        print(f"Could not calculate beta for {ticker}: {e}")

# Convert results to a DataFrame
beta_df = pd.DataFrame.from_dict(betas, orient='index', columns=['Beta'])

# import ace_tools as tools; tools.display_dataframe_to_user(name="S&P 500 Betas", dataframe=beta_df)
```

```python
# from IEX cloud
import yfinance as yf
import pandas as pd
from datetime import datetime
import requests

# Replace 'your_api_key' with your actual IEX Cloud API key
api_key = 'your_api_key'
url = f'https://cloud.iexapis.com/stable/ref-data/symbols?token={api_key}'

response = requests.get(url)
data = response.json()

# Extract S&P 500 symbols
sp500_tickers = [item['symbol'] for item in data if item['isEnabled'] and 'sp500' in item['tags']]

# Define the indices tickers
dow_jones_ticker = '^DJI'
nasdaq_ticker = '^IXIC'

# Define a start and end date for the data collection
start_date = '2020-01-01'
# end_date = datetime.today().strftime('%Y-%m-%d')
end_date = '2024-06-15'

# Initialize a list to store data
data = []

for ticker in sp500_tickers:
    stock = yf.Ticker(ticker)
    hist = stock.history(start=start_date, end=end_date)
    pe_ratio = stock.info.get('trailingPE', 'N/A')

    for date, row in hist.iterrows():
        data.append([
            ticker,
            date,
            row['Open'],
            row['Close'],
            row['Volume'],
            yf.Ticker(dow_jones_ticker).history(start=date, end=date)['Close'].iloc[0],
            yf.Ticker(nasdaq_ticker).history(start=date, end=date)['Close'].iloc[0],
            pe_ratio
        ])

# Create a DataFrame
df = pd.DataFrame(data, columns=[
    'Ticker', 'Date', 'Open Price', 'Close Price', 'Volume', 'Dow Jones Index', 'Nasdaq Index', 'P/E Ratio'
])

# Save the DataFrame to a CSV file
df.to_csv('sp500_data.csv', index=False)

print("Data saved to sp500_data.csv")
```
